```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
list = ['Preg', 'BP1', 'BP2', 'GL', 'IL', 'BMI', 'Skin Tone', 'Avg', 'Class']
```

```python
data = pd.read_csv('pima-indians-diabetes.csv', names = list)
```

```python
data.head(2)
```

```python
x = data.iloc[:, :-1]
x
```

```python
y = data.iloc[:, -1]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1)
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 10, activation = 'relu', input_dim = 8))
model.add(Dense(units = 8, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
model.get_weights()
```

```python
predictions = model.predict(x_test)
predictions
```

```python
class_label = []
```

```python
for i in range(predictions.size):
  if (predictions[i] > 0.51):
    class_label.append(1)
  else:
    class_label.append(0)
```

```python
accuracy_score = accuracy_score(class_label, y_test)
accuracy_score
```

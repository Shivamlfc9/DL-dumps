```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=100,noise=0.2,random_state=1)
```

```python
x.shape
```

```python
y
```

```python
n_train=30
```

```python
x_train,x_test=x[:n_train],x[n_train:]
```

```python
x_train
```

```python
x_train.shape
```

```python
x_test
```

```python
x_test.shape
```

```python
y_train,y_test=y[:n_train],y[n_train:]
```

```python
y_train
```

```python
y_train.shape
```

```python
y_test
```

```python
y_test.shape
```

```python
model=Sequential()
```

```python
model.add(Dense(500,input_dim=2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=40)
```

```python
plt.plot(history.history['accuracy'],label='train')
plt.plot(history.history['val_accuracy'],label='validation')
plt.legend()
```

```python
model_l2=Sequential()
```

```python
from keras.regularizers import l2
```

```python
model_l2.add(Dense(500,input_dim=2,activation='relu',kernel_regularizer=l2(0.001)))
```

```python
model_l2.add(Dense(1,activation='sigmoid'))
```

```python
model_l2.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
history_l2=model_l2.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=40)
```

```python
plt.plot(history_l2.history['accuracy'],label='train')
plt.plot(history_l2.history['val_accuracy'],label='validation')
plt.legend()
```
